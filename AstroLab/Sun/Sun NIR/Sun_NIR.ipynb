```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("SUN_INFRA0001.trt", delimiter=";", skiprows=8)
wavelength_source = data[:,0]
intensity_source = data[:,1]
```

```python
##Plot the data
plt.figure(figsize = (15,7.5))
plt.plot(wavelength_source, intensity_source, lw = 2)
plt.xlabel(r'Wavelength \lambda (nm)')
plt.title("Solar spectrum in near infrared regeion with noise")
plt.ylabel("Relative counts")
plt.grid()
plt.show()
```

```python
data = np.loadtxt("SUN_INFRA_BACKGROUND0001.trt", delimiter=";", skiprows=8)
wavelength_bg = data[:,0]
intensity_bg = data[:,1]
```

```python
##Plot the data
plt.figure(figsize = (15,7.5))
plt.plot(wavelength_bg, intensity_bg, lw = 2)
plt.xlabel(r'Wavelength \lambda (nm)')
plt.ylabel("Relative counts")
plt.title("Noise in the Solar spectrum in near infrared regeion")
plt.grid()
plt.show()
```

```python
data = intensity_source-intensity_bg
```

```python
plt.figure(dpi = 600, figsize = (15,7.5))
plt.plot(wavelength_source, intensity_source, linewidth = 1, color = 'r', label = "Data with noise")
plt.plot(wavelength_source, intensity_bg, linewidth = 1, color = 'g', label = "Noise")
plt.plot(wavelength_source, data, linewidth = 1, color = 'b', label = "Data without noise")
plt.xlabel(r"$\lambda$ (nm)")
plt.ylabel("Counts")
plt.legend(fontsize = 5)
plt.title("Sun Spectrum in Infrared Regeion without noise")
plt.grid(ls = "--")
plt.legend()
plt.show()
```
