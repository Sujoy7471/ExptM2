```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##Import Spectrum data
data = np.loadtxt("SUN_VIS0001.trt", delimiter=";", skiprows=7)                                                         ##Read data files
wavelength_source = data[:,0]
intensity_source = data[:,1]

##Plot the data
plt.figure(figsize = (20,10))
plt.plot(wavelength_source, intensity_source) 
plt.xlabel(r"Wavelength $\lambda$ (nm)")
plt.ylabel("Counts")
plt.title("Solar spectrum (With Noise)")
plt.grid()
plt.show()    
```

```python
##Import noise data
data = np.loadtxt("SUN_VIS0002.trt", delimiter=";", skiprows=7)                                                         ##Read noise file
wavelength_bg = data[:,0]
intensity_bg = data[:,1]

##Plot noise
plt.figure(figsize = (20,10))
plt.plot(wavelength_bg, intensity_bg)  
plt.xlabel(r"Wavelength $\lambda$ (nm)")
plt.ylabel("Counts")
plt.ylim([-20, 200])
plt.title("Noise in solar spectrum")
plt.grid()
plt.show()                     
```

```python
##Removing Noise from data
data = intensity_source-intensity_bg

##Plot the spectrum
plt.figure(figsize = (20,10))
plt.plot(wavelength_source, intensity_source, lw = 6, color = 'r', alpha = 0.5, label = "Data with Noise") 
plt.plot(wavelength_bg, intensity_bg, lw = 2, color = 'g', label = "Noise")  
plt.plot(wavelength_source, data, linewidth = 2, color = 'b', label = "Denoiced")
plt.xlabel(r"Wavelength $\lambda$ (nm)")
plt.ylabel("Counts")
plt.title("Solar spectrum (Without Noise)")
plt.legend()
plt.grid()
plt.show()    
```
