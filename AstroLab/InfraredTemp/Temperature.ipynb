```python
import numpy as np
import matplotlib.pyplot as plt

def temperature(lambda_1, lambda_2, V_1, V_2, delta_lambda_1, delta_lambda_2, n_1, n_2):
    h = 6.626e-34  # Planck's constant
    c = 3e8  # Speed of light
    kb = 1.38e-23  # Boltzmann constant

    # Compute the ratio C
    C = (n_2 * V_1 * lambda_1**5 * delta_lambda_2) / (n_1 * V_2 * lambda_2**5 * delta_lambda_1)

    # Temperature array
    T = np.arange(100, 2000, 0.1)
    
    # Compute y(T)
    y = (np.exp(h * c / (lambda_2 * kb * T)) - 1) / (np.exp(h * c / (lambda_1 * kb * T)) - 1)
    
    # Compute the absolute difference
    diff = np.abs(y - C)

    # Find the temperature corresponding to the minimum difference
    Tf = T[np.argmin(diff)]

    return Tf, T, diff

# Data format: [initial_lambda, final_lambda, n, signal]
J = [1.081e-6, 1.396e-6, 0.5, 1e-6]  # Small nonzero value instead of 0
H = [1.570e-6, 1.78e-6, 0.8, 0.030]
K = [2.041e-6, 2.41e-6, 0.7, 0.140]
L = [3.49e-6, 4.166e-6, 0.8, 1.380]
M = [4.424e-6, 5.012e-6, 0.8, 5.380]

data = [J, H, K, L, M]

plt.figure(figsize=(8, 6))

temperatures = []  # Store valid temperatures

# Loop through data pairs
for i in range(len(data)):
    for j in range(i+1, len(data)):  # Avoid duplicate calculations
        lambda_1 = (data[i][0] + data[i][1]) / 2
        delta_lambda_1 = (data[i][1] - data[i][0]) / 2
        n_1 = data[i][2]
        V_1 = data[i][3]
        
        lambda_2 = (data[j][0] + data[j][1]) / 2
        delta_lambda_2 = (data[j][1] - data[j][0]) / 2
        n_2 = data[j][2]
        V_2 = data[j][3]

        T, t_array, diff = temperature(lambda_1, lambda_2, V_1, V_2, delta_lambda_1, delta_lambda_2, n_1, n_2)

        temperatures.append(T)
        print(f"T for {i}-{j}: {T:.2f} K")

        plt.plot(t_array, diff, label=f"{i}-{j}")

# Compute and print the average temperature
avg_temperature = np.mean(temperatures)
print(f"\nAverage Temperature: {avg_temperature:.2f} K")

plt.xlabel("Temperature (K)")
plt.ylabel("Difference |y - C|")
plt.legend()
plt.title("Temperature Estimation")
plt.grid()
plt.show()
```
